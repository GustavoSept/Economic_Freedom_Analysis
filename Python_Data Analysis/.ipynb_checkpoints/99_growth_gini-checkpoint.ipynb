```python
def findNaN_Cols_OnYears(targetSeries, nameSeries, yearSeries, years):
    """
    Accepts:
        targetSeries: Series to check for NaN values
        nameSeries: Series of country names
        yearSeries: Series of years
        years: List of years to check for NaN values in targetSeries
    If there are NaN values in targetSeries for any of the specified years for a country, 
    that country is added to the countriesToDrop list.
    """
    
    countries_list = nameSeries.unique()
    countriesToDrop = []

    for country in countries_list: 
        for year in years:
            if targetSeries[(nameSeries == country) & (yearSeries == year)].isna().any(): 
                countriesToDrop.append(country)
                break # No need to check further years if one NaN is found

    return countriesToDrop

countries_to_remove = findNaN_Cols_OnYears(df_master['GDP per capita (current USD)'], df_master['Country Name'], df_master['Index Year'], [1995, 2010, 2021])


```

```python
def calculate_percent_difference(targetSeries, nameSeries, yearSeries, year_pair, in_countriesList = None):
    """
    Accepts:
        targetSeries: Series to compute the percent difference (e.g., GDP per capita)
        nameSeries: Series of country names
        yearSeries: Series of years
        year_pair: List of two years [start_year, end_year] to compute the percent difference

    Returns:
        A dictionary where keys are country names and values are the computed percent differences
        for the specified year pair.
    """
    if in_countriesList is None:
        countries_list = nameSeries.unique()
    else:
        countries_list = in_countriesList
    
    start_year, end_year = year_pair    
    percent_difference_dict = {}

    for country in countries_list:
        start_value = targetSeries[(nameSeries == country) & (yearSeries == start_year)].values
        end_value = targetSeries[(nameSeries == country) & (yearSeries == end_year)].values

        # Check if both start and end values exist and are not NaN
        if len(start_value) > 0 and len(end_value) > 0 and not np.isnan(start_value) and not np.isnan(end_value):
            difference = ((end_value - start_value) / start_value) * 100
            percent_difference_dict[country] = difference[0]  # Extract the scalar value from the array

    return percent_difference_dict
```
